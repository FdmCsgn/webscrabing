```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
r=requests.get("https://www.woah.org/en/what-we-do/animal-health-and-welfare/animal-diseases/?_paged=14")
soup = BeautifulSoup(r.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
hastalıklar=soup.find_all("h3",attrs={"class":"cards__title"})
```

```python
hastalıklar
```

```python
for hastalık in hastalıklar:
    # <a> etiketlerini bul ve 'title' niteliğini al
    linkler = hastalık.find_all("a", attrs={"title": True})  # 'title' niteliği olan <a> etiketlerini bul
    for link in linkler:
        title = link.get("title")  # 'title' niteliğini al
        print(title)


    




    
```
